```python
import pandas as pd
from tqdm import tqdm

# Enable tqdm for pandas
tqdm.pandas()

# Read CSV in chunks with a progress bar
chunksize = 100_000  # Adjust this based on your system's memory
filename = '/Users/deberend/Desktop/Cornell Tech/Applied Machine Learning/Final_Project/data/LSEG_full_dataset.csv'

# Initialize an empty list to store processed chunks
data = []

for chunk in tqdm(pd.read_csv(filename, chunksize=chunksize)):
    # Optional: Process the chunk here (e.g., filtering or transformations)
    data.append(chunk)

# Combine all chunks into a single DataFrame
df = pd.concat(data, ignore_index=True)
```
