```python
def prepare_panel_dataset(df, groupby=True):
    """
    This function makes it so that the data is in a panel setting
    Indexed on ISIN instead of ticker as ISIN's are unique
    """
    
    # Clean column names (remove spaces, standardize)
    df.columns = df.columns.str.strip().str.replace(' ', '_')
    
    # Ensure year and isin are properly identified
    df['year'] = df['year'].astype(int)
    df['isin'] = df['isin'].astype(str)
    
    # Group by and sort by year if required
    if groupby:
        df = df.groupby('isin').apply(lambda x: x.sort_values('year')).reset_index(drop=True)
    
    # Set MultiIndex for panel data analysis
    df_panel = df.set_index(['isin', 'year'])
    
    return df_panel
```

```python
df_panel = prepare_panel_dataset(pivoted_data)
display(df_panel)
```

```python
panel_dataset = df_panel.copy()
```

# Now we have an initial preprocessed dataset

```python
import pandas as pd

def preprocess_data(raw_df, pivot=True, filter_missing_threshold=10, impute_missing=True):
    """
    Preprocess the S&P 500 dataset.

    Parameters:
        raw_df (pd.DataFrame): The raw dataset.
        pivot (bool): Whether to pivot the dataset from stacked format to a panel format.
        filter_missing_threshold (int): The threshold (%) for filtering out features with too much missing data.
        impute_missing (bool): Whether to impute missing data using forward and backward fill.

    Returns:
        pd.DataFrame: The preprocessed dataset.
    """
    # Step 1: Pivot the dataset
    if pivot:
        df = raw_df.pivot_table(
            index=['orgpermid', 'year', 'isin', 'ticker', 'comname'],
            columns='fieldname',
            values='valuescore',
            aggfunc='first'
        ).reset_index()

        # Flatten the MultiIndex columns
        df.columns.name = None
        df.columns = [str(col) for col in df.columns]
    else:
        df = raw_df.copy()

    # Step 2: Prepare panel dataset
    def prepare_panel_dataset(df, groupby=True):
        df.columns = df.columns.str.strip().str.replace(' ', '_')
        df['year'] = df['year'].astype(int)
        df['isin'] = df['isin'].astype(str)
        if groupby:
            df = df.groupby('isin').apply(lambda x: x.sort_values('year')).reset_index(drop=True)
        df_panel = df.set_index(['isin', 'year'])
        return df_panel

    df = prepare_panel_dataset(df)

    # Step 3: Filter out features with too much missing data
    if filter_missing_threshold is not None:
        missing_percentages = df.isna().mean() * 100
        filtered_features = missing_percentages[missing_percentages <= filter_missing_threshold].index
        df = df[filtered_features]

    # Step 4: Impute missing data
    if impute_missing:
        df = df.groupby(level='isin').ffill().bfill()

    return df
```

```python
# Test the preprocessing function:
df_initial_preprocess = preprocess_data(sp500_data, pivot=True, filter_missing_threshold=10, impute_missing=True)
display(df_initial_preprocess)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Drop non-feature columns
pca_input_data = df_initial_preprocess.drop(columns=['orgpermid', 'ticker', 'comname'])

# Step 2: Standardize the data (PCA requires features to be on the same scale)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(pca_input_data)

# Step 3: Apply PCA
pca = PCA()
pca_results = pca.fit_transform(scaled_data)

# Convert PCA results to a DataFrame for better interpretability
pca_df = pd.DataFrame(pca_results, index=pca_input_data.index, 
                      columns=[f'PC{i+1}' for i in range(pca_results.shape[1])])
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Cumulative Explained Variance Scree Plot
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance Ratio')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.ylim(0, 1)  # Set y-axis from 0 to 1 (0% to 100%)
plt.grid(True)
plt.show()
```
